```python
##DuckDuckGo Internet Search with robots.txt check for permissions to scrape
#search_and_scrape(query, return_number=5, pdf_dir="pdfs") : function for querying DDG internet search, webscraping if permitted; if return is pdf, handle pdf
#|can_scrape_url(url, user_agent='*') : function for checking if we can scrape the returned site or not
#|download_pdf(url, save_path=None) :
#|pdf_to_text() : convert pdf to single text str using fitz library

import requests
from bs4 import BeautifulSoup
from urllib.parse import quote_plus, urlparse, urljoin, unquote
from urllib.robotparser import RobotFileParser
import time
import os
import fitz

def can_scrape_url(url, user_agent='*'):
    try:
        parsed_url = urlparse(url)
        base_url = f"{parsed_url.scheme}://{parsed_url.netloc}"

        rp = RobotFileParser()
        robots_url = urljoin(base_url, '/robots.txt')
        rp.set_url(robots_url)

        try:
            rp.read(timeout=5)
        except Exception:
            return True

        return rp.can_fetch(user_agent, url)
    except Exception as e:
        print(f"Robots.txt check error: {e}")
        return True

def download_pdf(url, save_path=None):
    try:
        response = requests.get(url)
        response.raise_for_status()
        content = response.content

        if 'Content-Disposition' in response.headers:
            cd = response.headers['Content-Disposition']
            if 'filename=' in cd:
                filename = cd.split('filename=')[1].strip('"\'')
            else:
                filename = 'downloaded.pdf'
        else:
            filename = unquote(url.split('/')[-1])
            if not filename.endswith('.pdf'):
                filename = 'downloaded.pdf'

        if save_path is None:
            save_path = os.getcwd()

        if os.path.isdir(save_path) or save_path.endswith('/') or save_path.endswith('\\'):
            full_path = os.path.join(save_path, filename)
        else:
            full_path = save_path

        os.makedirs(os.path.dirname(os.path.abspath(full_path)), exist_ok=True)

        with open(full_path, 'wb') as f:
            f.write(content)

        return full_path

    except Exception as e:
        print(f"Error downloading PDF: {e}")
        return None

def pdf_to_text(pdf_path):
    try:
        doc = fitz.open(pdf_path)
        text = ""
        for page in doc:
            text += page.get_text("text") + "\n"
        doc.close()
        return text.strip()
    except Exception as e:
        print(f"Error converting PDF: {e}")
        return None

def search_and_scrape(query, return_number=5, pdf_dir='pdfs'):
    """Function for querying DDG internet search, webscraping if permitted; if return is pdf, handle pdf"""
    result_dict = {}
    successful_scrapes = 0
    user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    headers = {'User-Agent': user_agent}

    try:
        search_url = f"https://duckduckgo.com/html/?q={quote_plus(query)}"
        print(f"Searching for: {query}")

        time.sleep(1)
        search_response = requests.get(search_url, headers=headers)
        search_response.raise_for_status()

        soup = BeautifulSoup(search_response.text, 'html.parser')
        results = soup.find_all('div', class_='result__body')
        urls = [a['href'] for result in results for a in result.find_all('a', class_='result__url')]

        print(f"Found {len(urls)} potential URLs")

        for url in urls:
            if successful_scrapes >= return_number:
                break

            if not url:
                continue

            print(f"Attempting URL {successful_scrapes + 1}/{return_number}: {url}")

            if not can_scrape_url(url, user_agent):
                print(f"Skipping prohibited URL: {url}")
                continue

            try:
                time.sleep(2)

                # Check if URL is a PDF
                is_pdf = url.lower().endswith('.pdf') or (
                    requests.head(url, headers=headers, allow_redirects=True).headers.get('content-type', '')
                    .lower().startswith('application/pdf')
                )

                if is_pdf:
                    print(f"Processing PDF: {url}")
                    pdf_path = download_pdf(url, pdf_dir)
                    if pdf_path:
                        text = pdf_to_text(pdf_path)
                        if text:
                            result_dict[url] = text[:5000]
                            successful_scrapes += 1
                            print(f"Successfully processed PDF: {url}")
                            try:
                                os.remove(pdf_path)  # Clean up PDF file
                            except:
                                pass
                else:
                    page_response = requests.get(url, headers=headers, timeout=10)
                    page_response.raise_for_status()

                    soup = BeautifulSoup(page_response.text, 'html.parser')

                    for element in soup(['script', 'style', 'nav', 'footer', 'header', 'noscript']):
                        element.decompose()

                    text = soup.get_text(separator='\n', strip=True)
                    text = '\n'.join(line.strip() for line in text.split('\n') if line.strip())

                    result_dict[url] = text[:5000]
                    successful_scrapes += 1
                    print(f"Successfully scraped HTML: {url}")

            except Exception as e:
                print(f"Error processing {url}: {str(e)[:100]}...")
                continue

    except Exception as e:
        print(f"Critical error: {str(e)}")

    print(f"\nCompleted with {successful_scrapes} successful scrapes")
    return result_dict

#query = "farming best practices filetype:pdf"  # Add filetype:pdf to specifically search for PDFs
query = "sustainable biodiversity soil management"
result_dict = search_and_scrape(query, return_number=3, pdf_dir='.')

for url, content in result_dict.items():
    print(f"\nSource: {url}")
    print("Content preview:", content[:200], "...\n")
    print("-" * 80)
```

```python
#split_observation(observation, source="unknown") : split `observation` full text string into chunked observations using RecursiveCharacterTextSplitter(); track 'source' of the information; return dict with key per observation chunk and key per source (assumed 1 source)
#

import requests
import fitz  # PyMuPDF
import io
import os
from urllib.parse import unquote
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_observation(observation, source="unknown", chunk_size=500,chunk_overlap=150):
    """
    Inputs: `observation`: full text str observed, `source`: str
    """
    # Initialize the splitter
    splitter = RecursiveCharacterTextSplitter(
        # chunk_size=700,
        # chunk_overlap=150,
        # length_function=len,
        # is_separator_regex=False,
      chunk_size=chunk_size,  # Optimal for 2048 ctx window
      chunk_overlap=chunk_overlap,
      separators=["\n\n", ". ", "! ", "? "]  # Entity-preserving splits
      )

    # Split the observation into chunks
    chunks = splitter.split_text(observation)

    # Create the observation_list dictionary
    observation_dict = {
        "observations": chunks,
        "source": source
    }

    return observation_dict


url = "https://journals.plos.org/climate/article/file?id=10.1371/journal.pclm.0000169&type=printable"

print("File downloaded")
def download_and_extract_pdf_text(url, save_path=None):
    pdf_content = download_pdf(url, save_path=save_path)
    text = pdf_to_text(pdf_content)
    if text:
        print(f"Text length = {len(text)} characters")
        print(text[30000:31000])  # Print first 500 characters
    else:
        print("Error: No text extracted from PDF")
    return text

text = download_and_extract_pdf_text(url, save_path=None)

#observation_dict_crops = split_observation(observation, source="327_NH_GD_(Con)servation_Crops-Seeding_Warm_Season_Grasses_2011")
observation_dict_biofinance = split_observation(text, source="journal.pclm.0000169.pdf")

url2 = "https://arxiv.org/pdf/1301.5968"
text2 = download_and_extract_pdf_text(url2, save_path=None)
observation_dict_biomaterials1 = split_observation(text2, source="https://arxiv.org/pdf/1301.5968")
# print(observation_dict_crops)
# print(f"len(observation_dict_crops = {len(observation_dict_crops['observations'])}")
print(f"len(observation_dict_biofinance = {len(observation_dict_biofinance['observations'])}")
for key in observation_dict_biofinance.keys():
  print(f"{key} = {observation_dict_biofinance[key]}")
```

```python
#Arxiv API
# retrieve_papers(query, return_number, download_path) : send query to arxiv API to return `return_number` top matches and download

import arxiv
import fitz
import os
from typing import Dict

def retrieve_papers(query: str, return_number: int = 2, download_path: str = '.') -> Dict[str, str]:
    """
    Retrieves arXiv papers matching a query, downloads PDFs, and extracts full text.

    Args:
        query: Search string for arXiv papers
        return_number: Number of papers to retrieve (default: 2)
        download_path: Directory to save PDFs (default: current directory)

    Returns:
        Dictionary with citations as keys and extracted text as values
    """

    # Create client with rate limiting
    client = arxiv.Client(
        page_size=100,
        delay_seconds=3,
        num_retries=3
    )

    # Configure search parameters
    search = arxiv.Search(
        query=query,
        max_results=return_number,
        sort_by=arxiv.SortCriterion.Relevance,
        sort_order=arxiv.SortOrder.Descending
    )

    result_dict = {}
    citations = []

    try:
        # Process search results
        for paper in client.results(search):
            # Generate citation
            authors = [a.name for a in paper.authors]
            citation = (
                f"{', '.join(authors[:2])} et al. "
                f"'{paper.title}', "
                f"arXiv:{paper.entry_id.split('/')[-1]}, "
                f"{paper.published.strftime('%d %b %Y')}"
            )

            # Store citation
            citations.append(citation)
            print(f"Processing: {citation}")

            # Download PDF
            safe_title = ''.join(c if c.isalnum() else '_' for c in paper.title)
            filename = f"{safe_title[:50]}_{paper.get_short_id()}.pdf"
            filepath = os.path.join(download_path, filename)

            paper.download_pdf(dirpath=download_path, filename=filename)

            # Extract text from PDF
            doc = fitz.open(filepath)
            full_text = []
            for page in doc:
                text = page.get_text("text")
                full_text.append(text.strip())

            # Clean up temporary file
            doc.close()
            os.remove(filepath)

            # Store in result dictionary
            result_dict[citation] = '\n\n'.join(full_text)

    except Exception as e:
        print(f"Error processing paper: {str(e)}")

    return result_dict
```

```python
# extraction_information_test() with new validation to create missing nodes (0.5 probability and description stating auto-created)
from pydantic import BaseModel, field_validator
from typing import Dict, List
from pydantic_core.core_schema import FieldValidationInfo
from ollama import chat
import datetime

class KeyConcept(BaseModel):
    description: str
    probability: float
    attributes: Dict = {}

class Relationship(BaseModel):
    source: str
    target: str
    type: str
    probability: float
    description: str

class ExtractedDataNoValidator(BaseModel):
    key_concepts: Dict[str, KeyConcept]
    relationships: List[Relationship]

class ExtractedDataValidator(BaseModel):
    key_concepts: Dict[str, KeyConcept]
    relationships: List[Relationship]

    @field_validator('relationships')
    @classmethod
    def validate_relationship_nodes(cls, v, info: FieldValidationInfo):
        if 'key_concepts' not in info.data:
            raise ValueError('key_concepts must be provided before relationships')

        key_concepts = info.data['key_concepts']

        # Create a mutable copy of key_concepts that we can modify
        key_concepts_mutable = dict(key_concepts)

        # For each relationship, create missing nodes
        for rel in v:
            # Check and create source node if missing
            if rel.source not in key_concepts_mutable:
                key_concepts_mutable[rel.source] = KeyConcept(
                    description=f"Auto-created from relationship: {rel.source}",
                    probability=0.5,  # Default probability for auto-created nodes
                    attributes={}
                )

            # Check and create target node if missing
            if rel.target not in key_concepts_mutable:
                key_concepts_mutable[rel.target] = KeyConcept(
                    description=f"Auto-created from relationship: {rel.target}",
                    probability=0.5,  # Default probability for auto-created nodes
                    attributes={}
                )

        # Update the original key_concepts with new nodes
        info.data['key_concepts'].update(key_concepts_mutable)
        return v

import signal
from functools import wraps
from typing import Dict, List, Optional, Callable, Any

def timeout_handler(signum, frame):
    raise TimeoutError()

def with_timeout(timeout_seconds: Optional[float]) -> Callable:
    def decorator(func: Callable) -> Callable:
        @wraps(func)
        def wrapper(*args, **kwargs) -> Any:
            if timeout_seconds is None:
                return func(*args, **kwargs)

            # Set the signal handler
            signal.signal(signal.SIGALRM, timeout_handler)
            signal.alarm(int(timeout_seconds))  # Schedule the alarm

            try:
                result = func(*args, **kwargs)
            finally:
                signal.alarm(0)  # Disable the alarm

            return result
        return wrapper
    return decorator

def extract_information_test(
    observation: str,
    model_name,
    source="unknown",
    timestamp="",
    goal_prior="",
    timeout_seconds: Optional[float] = None
) -> Dict:
    """Extract structured information from observation with optional timeout"""
    print(f"\nExtracting information from: {observation[:100]}...")

    try:
        @with_timeout(timeout_seconds)
        def _extract_with_timeout() -> Dict:
            sys_prompt = """Extract ALL key concepts and relationships in JSON format and assign a probability value denoting how confident you are in the truth of your description. The "source" and "target" for each of the "relationships" must be a concept in "key_concepts".
            For each statement or implied relationship in the text, create a separate relationship entry.
            Return ONLY the JSON with no additional text or markdown formatting:
            {
                "key_concepts": {
                    "concept1": {
                        "description": "description of concept1",
                        "probability": 0.8,
                        "attributes": {}
                    },
                    "concept2": {
                        "description": "description of concept2",
                        "probability": 0.8,
                        "attributes": {}
                    }
                },
                "relationships": [
                    {
                        "source": "concept1",
                        "target": "concept2",
                        "type": "relationship_type",
                        "probability": 0.7,
                        "description": "description of relationship between concept1 and concept2"
                    }
                ]
            }"""

            user_prompt = f"{goal_prior} Extract ALL key concepts and relationships from the following: {observation}"

            print("Sending request to LLM...")
            response = chat(
                model=model_name,
                messages=[
                    {"role": "system", "content": sys_prompt},
                    {"role": "user", "content": user_prompt}
                ],
                format=ExtractedDataValidator.model_json_schema(),
                options={
                        "temperature": 0,        # Controls randomness (0 = deterministic)
                        "timeout": timeout_seconds,  # Maximum time to wait for response
                        #"num_gpu": 0, #1,           # Number of GPU layers to use
                        "num_ctx": 1024, #2048,        # Context window size in tokens
                        "top_p": 0.1, #0.9,           # sampling threshold
                        "num_thread": 4,        # CPU threads for computation (4 supposedly best for T4 GPU)
                        "repeat_penalty": 1.1   # Penalizes repetition (>1 reduces repeats)
                        #"keep_alive": 5         # Seconds to keep model loaded in memory
                    }
            )

            # Parse JSON using Pydantic for validation
            parsed_data = ExtractedDataValidator.model_validate_json(response['message']['content'])
            data = parsed_data.model_dump()

            # Add metadata
            current_time = timestamp or datetime.datetime.now().isoformat()
            metadata = {
                'timestamp': current_time,
                'source_document': source,
                'source_observation': observation,
                'model_name': model_name,
                'goal_prior': goal_prior
            }

            # Add metadata to key_concepts and relationships
            for concept in data['key_concepts'].values():
                concept.update(metadata)
            for relationship in data['relationships']:
                relationship.update(metadata)

            return data

        return _extract_with_timeout()

    except TimeoutError:
        print(f"Error: Timeout after {timeout_seconds} seconds")
        return {"key_concepts": {}, "relationships": []}
    except Exception as e:
        print(f"Error in extract_information_test: {type(e).__name__}: {e}")
        if not model_name:
            print(f"Error: No model name specified.")
        return {"key_concepts": {}, "relationships": []}


```

```python
# update_probability_with_metrics(prior, new_prob) : apply free energy minimization rule in function for graph node/edge probability updating for sequential observations
#

from scipy import optimize
import numpy as np

def update_probability_with_metrics(prior, new_prob):
    """
    Update probability and return all metrics from free energy minimization for a graph node/edge.
    """
    def free_energy(p_candidate):
        complexity = (p_candidate - prior)**2
        accuracy = (p_candidate - new_prob)**2

        epsilon = 1e-10
        entropy = -(p_candidate * np.log(p_candidate + epsilon) +
                   (1-p_candidate) * np.log(1-p_candidate + epsilon))

        return complexity + accuracy - 0.1 * entropy, complexity, accuracy

    result = optimize.minimize_scalar(
        lambda x: free_energy(x)[0],
        bounds=(0, 1),
        method='bounded'
    )

    final_free_energy, final_complexity, final_accuracy = free_energy(result.x)
    return result.x, final_free_energy, final_complexity, final_accuracy

def consolidate_knowledge_graphs(observation_graphs, mode="average"):
    """Consolidate observation graphs into a single graph sequentially."""
    consolidated = {
        'key_concepts': {},
        'relationships': []
    }

    # Process key concepts (nodes)
    for graph in observation_graphs:
        if not graph['key_concepts']:
            continue

        for concept, details in graph['key_concepts'].items():
            if not details['description']:
                continue

            if concept in consolidated['key_concepts']:
                # Combine existing node
                existing = consolidated['key_concepts'][concept]

                # Update probability and get metrics
                if mode == "average":
                    new_prob = (existing['probability'] + details['probability']) / 2
                    free_energy_value = float('inf')
                    complexity_value = float('inf')
                    accuracy_value = 0.0
                else:
                    new_prob, free_energy_value, complexity_value, accuracy_value = update_probability_with_metrics(
                        existing['probability'],
                        details['probability']
                    )

                # Update probability
                existing['probability'] = new_prob

                # Update description
                if details['description'] and details['description'] not in existing['description']:
                    existing['description'] = (existing['description'] or '') + ('; ' + details['description'] if existing['description'] else details['description'])

                # Merge attributes
                existing['attributes'].update(details['attributes'])

                # Update history lists
                for field in ['timestamp', 'source', 'source_observation', 'model_name']:
                    if field not in existing:
                        existing[field] = []
                    if isinstance(existing[field], str):
                        existing[field] = [existing[field]]
                    existing[field].append(details.get(field, ''))

                # Update metric histories
                if 'probability_history' not in existing:
                    existing['probability_history'] = [existing['probability']]
                existing['probability_history'].append(new_prob)

                if 'free_energy_history' not in existing:
                    existing['free_energy_history'] = [float('inf')]
                existing['free_energy_history'].append(free_energy_value)

                if 'complexity_history' not in existing:
                    existing['complexity_history'] = [float('inf')]
                existing['complexity_history'].append(complexity_value)

                if 'accuracy_history' not in existing:
                    existing['accuracy_history'] = [0.0]
                existing['accuracy_history'].append(accuracy_value)

                if 'goal_prior_history' not in existing:
                    existing['goal_prior_history'] = [existing.get('goal_prior', '')]
                existing['goal_prior_history'].append(details.get('goal_prior', ''))

                if 'timestamp_history' not in existing:
                    existing['timestamp_history'] = [existing.get('timestamp', '')]
                existing['timestamp_history'].append(details.get('timestamp', ''))

                if 'description_history' not in existing:
                    existing['description_history'] = [existing.get('description', '')]
                existing['description_history'].append(details.get('description',''))

                #if 'summary_history' not in existing:
                #    existing['summary_history'] = [existing.get('summary', '')]
                #existing['description_history'].append(details.get('summary', '')

            else:
                # Add new node
                new_node = {
                    'description': details['description'],
                    'probability': details['probability'],
                    'attributes': details['attributes'].copy(),
                    'probability_history': [details['probability']],
                    'free_energy_history': [float('inf')],
                    'complexity_history': [float('inf')],
                    'accuracy_history': [0.0],
                    'goal_prior_history': [details.get('goal_prior', '')],
                    'timestamp_history': [details.get('timestamp', '')],
                    'description_history': [details.get('description','')]
                }

                # Initialize history lists
                for field in ['timestamp', 'source', 'source_observation', 'model_name']:
                    new_node[field] = [details.get(field, '')]

                consolidated['key_concepts'][concept] = new_node

    # Process relationships (edges)
    seen_relationships = {}
    for graph in observation_graphs:
        if not graph['relationships']:
            continue

        for rel in graph['relationships']:
            if not rel['description']:
                continue

            rel_key = (rel['source'], rel['target'], rel['type'])
            rel_probability = rel.get('probability', 0.0)

            if rel_key in seen_relationships:
                idx = seen_relationships[rel_key]
                existing_rel = consolidated['relationships'][idx]

                # Update probability and get metrics
                if mode == "average":
                    new_prob = (existing_rel['probability'] + rel_probability) / 2
                    free_energy_value = float('inf')
                    complexity_value = float('inf')
                    accuracy_value = 0.0
                else:
                    new_prob, free_energy_value, complexity_value, accuracy_value = update_probability_with_metrics(
                        existing_rel['probability'],
                        rel_probability
                    )

                # Update probability
                existing_rel['probability'] = new_prob

                # Update description
                if rel['description'] and rel['description'] not in existing_rel['description']:
                    existing_rel['description'] = (existing_rel['description'] or '') + ('; ' + rel['description'] if existing_rel['description'] else rel['description'])

                # Update history lists
                for field in ['timestamp', 'source_document', 'source_observation', 'model_name']:
                    if field not in existing_rel:
                        existing_rel[field] = []
                    if isinstance(existing_rel[field], str):
                        existing_rel[field] = [existing_rel[field]]
                    existing_rel[field].append(rel.get(field, ''))

                # Update metric histories
                if 'probability_history' not in existing_rel:
                    existing_rel['probability_history'] = [existing_rel['probability']]
                existing_rel['probability_history'].append(new_prob)

                if 'free_energy_history' not in existing_rel:
                    existing_rel['free_energy_history'] = [float('inf')]
                existing_rel['free_energy_history'].append(free_energy_value)

                if 'complexity_history' not in existing_rel:
                    existing_rel['complexity_history'] = [float('inf')]
                existing_rel['complexity_history'].append(complexity_value)

                if 'accuracy_history' not in existing_rel:
                    existing_rel['accuracy_history'] = [0.0]
                existing_rel['accuracy_history'].append(accuracy_value)

                if 'goal_prior_history' not in existing_rel:
                    existing_rel['goal_prior_history'] = [existing_rel.get('goal_prior', '')]
                existing_rel['goal_prior_history'].append(rel.get('goal_prior', ''))

                if 'timestamp_history' not in existing_rel:
                    existing_rel['timestamp_history'] = [existing_rel.get('timestamp', '')]
                existing_rel['timestamp_history'].append(rel.get('timestamp', ''))

                if 'description_history' not in existing_rel:
                    existing_rel['description_history'] = [existing_rel.get('description', '')]
                existing_rel['description_history'].append(rel.get('description',''))

                #if 'summary_history' not in existing_rel:
                #    existing_rel['summary_history'] = [existing_rel.get('summary', '')]
                #existing_rel['description_history'].append(rel.get('summary', '')

            else:
                # Add new relationship
                new_rel = {
                    'source': rel['source'],
                    'target': rel['target'],
                    'type': rel['type'],
                    'probability': rel_probability,
                    'description': rel['description'],
                    'probability_history': [rel_probability],
                    'free_energy_history': [float('inf')],
                    'complexity_history': [float('inf')],
                    'accuracy_history': [0.0],
                    'goal_prior_history': [rel.get('goal_prior', '')],
                    'timestamp_history': [rel.get('timestamp', '')],
                    'description_history': [rel['description']]
                }

                # Initialize history lists
                for field in ['timestamp', 'source', 'source_observation', 'model_name']:
                    new_rel[field] = [rel.get(field, '')]

                consolidated['relationships'].append(new_rel)
                seen_relationships[rel_key] = len(consolidated['relationships']) - 1

    return consolidated


consolidated_graph = consolidate_knowledge_graphs(observation_graphs, mode="free_energy")
```

```python
# In progress:
# compute_graph_metrics for a graph
# enhance_semantic_connections

import networkx as nx
from networkx.algorithms.centrality import betweenness_centrality, eigenvector_centrality
from networkx.algorithms.link_analysis.pagerank_alg import pagerank

def compute_graph_metrics(graph):
    """
    Compute advanced graph metrics for nodes and edges in a knowledge graph.
    Returns node_metrics and edge_metrics dictionaries.
    """
    key_concepts = graph['key_concepts']
    relationships = graph['relationships']

    G = nx.DiGraph()

    # Add nodes with attributes
    for concept, details in key_concepts.items():
        G.add_node(concept, **details)

    # Add edges with attributes
    for rel in relationships:
        source = rel['source'][0] if isinstance(rel['source'], list) else rel['source']
        target = rel['target']

        if source not in G:
            G.add_node(source, probability=0.0, description="")
        if target not in G:
            G.add_node(target, probability=0.0, description="")

        G.add_edge(source, target, **rel)

    # Node metrics
    node_metrics = {
        node: {
            'degree': G.degree(node),
            'page_rank': pagerank(G).get(node, 0),
            'bridging_coefficient': betweenness_centrality(G).get(node, 0),
            'domain_prestige': eigenvector_centrality(G).get(node, 0)
        }
        for node in G.nodes()
    }

    # Edge metrics
    edge_metrics = {
        (u, v): {
            'weight': data.get('weight', 1.0),
            'probability': data.get('probability', 0.5),
            'type': data.get('type', 'undefined'),
            'centrality': betweenness_centrality(G, normalized=True, seed=42).get((u, v), 0)
        }
        for u, v, data in G.edges(data=True)
    }

    return node_metrics, edge_metrics

def enhance_semantic_connections(graph, similarity_threshold=0.85):
    """
    Enhance graph with semantic relationships using transformer embeddings.
    Returns updated graph with new semantic edges added to relationships.
    """
    key_concepts = graph['key_concepts']
    relationships = graph['relationships']

    try:
        from sentence_transformers import SentenceTransformer
        from sklearn.metrics.pairwise import cosine_similarity
    except ImportError:
        raise ImportError("Please install required packages: pip install sentence-transformers scikit-learn")

    # Load lightweight transformer model
    model = SentenceTransformer('all-MiniLM-L6-v2')

    # Extract and encode descriptions
    node_descriptions = {node: data['description'] for node, data in key_concepts.items()}
    embeddings = model.encode(list(node_descriptions.values()))

    # Compute semantic similarity matrix
    sim_matrix = cosine_similarity(embeddings)
    nodes = list(node_descriptions.keys())

    # Add new semantic relationships
    new_relationships = []
    for i in range(len(nodes)):
        for j in range(i+1, len(nodes)):
            if sim_matrix[i][j] >= similarity_threshold:
                new_relationships.append({
                    'source': nodes[i],
                    'target': nodes[j],
                    'type': 'SEMANTIC_SIMILARITY',
                    'probability': float(sim_matrix[i][j]),
                    'description': f"Semantic similarity: {sim_matrix[i][j]:.2f}",
                    'weight': float(sim_matrix[i][j])
                })

    # Return modified graph with updated relationships
    return {
        'key_concepts': key_concepts,
        'relationships': relationships + new_relationships
    }
```

```python
# extract_sentences(filename) to extract sentences from a .txt file

import spacy
import re

def extract_sentences(filename):
    # Load English language model
    nlp = spacy.load("en_core_web_sm")
    sentences = []

    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()

        # Split text into chunks by markdown headers
        chunks = re.split(r'((?:^|\n)#{1,6}\s+.*?)(?=\n|$)', text)

        for chunk in chunks:
            # Skip empty chunks
            if not chunk.strip():
                continue

            # If chunk is a header, add it directly
            if chunk.strip().startswith('#'):
                sentences.append(chunk.strip())
                continue

            # Process regular text with spaCy
            doc = nlp(chunk)
            for sent in doc.sents:
                # Clean and add each sentence
                cleaned = ' '.join(str(sent).split())
                if cleaned:
                    sentences.append(cleaned)

    return sentences


# # Example usage
# filename = f"{inputdirectory}/ma_hampshire_business_plan.txt"
# sentence_list = extract_sentences(filename)

# # Print each sentence
# for i, sentence in enumerate(sentence_list, 1):
#     print(f"Sentence {i}: {sentence}")
```
